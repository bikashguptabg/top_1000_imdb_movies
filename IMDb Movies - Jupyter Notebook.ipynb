## Problem Statement
We are an entertainment critic company called **'Kritics'** aiming to publish insightful facts about the most famous movies worldwide. Various stakeholders can utilize this dataset for different purposes:

- **'General public**': The data helps movie enthusiasts discover movies that match their preferences.
- **'Directors, producers, and other industry professionals**': The insights provide an understanding of how factors such as a movie's runtime, certificate category, and other variables predict its success, particularly in terms of IMDB rating and gross revenue.
- The key question guiding this project is:  
How can a movie's runtime, certificate category, and other attributes influence its success?

## Dataset Summary
The IMDB Top 1000 Movies Dataset is a comprehensive collection of information about the top 1000 highest-rated movies and TV shows as ranked by IMDb users between the period of these 100 years (1920-2020). This dataset contains 16 key attributes, including the title, release year, genre, runtime, and IMDB rating of each film. It also includes details like the director, leading stars, number of votes, and gross earnings. By analyzing this dataset, we can explore various trends in the film industry, such as popular genres, top-rated directors, and the correlation between ratings and revenue.

## Column Description in the IMDB Dataset

**1. Poster_Link**: The URL link to the movie or TV show's poster image, useful for visual representation.

**2. Series_Title**: The name or title of the movie or TV show.

**3. Released_Year**: The year in which the movie was released or the TV show first aired.

**4. Certificate**: The age rating (e.g., PG, R), indicating the appropriate audience for the content.
- **U:** Suitable for all ages.  
- **A:** Adults (18+), contains strong language/violence.  
- **UA:** 12+, parental guidance for younger viewers.  
- **R:** 17+, adult content.  
- **PG-13:** 13+, may not be suitable for children.  
- **PG:** Parental guidance suggested, suitable for all.  
- **Passed/Approved:** General audiences, old terms.  
- **G:** Suitable for all ages.  
- **TV-PG:** Parental guidance advised.  
- **GP:** Early term, like PG.  
- **TV-14:** Ages 14+.  
- **16:** 16+, common in Europe.  
- **TV-MA:** 17+, mature content.  
- **Unrated:** Not submitted for rating.  
- **U/A:** Same as UA, used in some regions.

**5. Runtime**: The total duration of the movie or the average duration of an episode, typically measured in minutes.

**6. Genre**: The category or type of the movie or TV show (e.g., Action, Drama, Comedy).

**7. IMDB_Rating**: The rating given by IMDb users, representing the audience's perception of the movie or show, typically on a scale of 1 to 10.

**8. Overview**: A brief synopsis or summary of the movie or TV showâ€™s plot.

**9. Meta_score**: A critic-based score, often aggregated from major film reviews. It reflects professional opinion and ranges from 28 to 100 in this dataset.

**10. Director**: The name of the director who directed the movie or TV show.

**11. Star1**: The lead actor or actress of the movie or TV show.

**12. Star2**: The second main actor or actress.

**13. Star3**: The third main actor or actress.

**14. Star4**: The fourth main actor or actress.

**15. No_of_Votes**: The total number of votes received by the movie or TV show on IMDb, indicating its popularity.

**16. Gross**: The total gross earnings of the movie or TV show, representing its financial success. Note that these figures are only for the USA and Canada, not worldwide earnings.

### Problem 3
- Incorrect dtypes for Release_Year, Runtime, Genre, Meta_score and Gross
#### Define
- **Changing the dtypes** : Released_Year (int32), Runtime (int16), Genre (category), Meta_score (int8), Gross (int64)

### Problem 7
- **Bucketing of `Certicate` columns** as there are many similar types of certifications that can be grouped.
#### Define
- Classification under these categories with a new column named 'Certificate_Category':
    - **For General Audiences (All Ages)**: U, G, Approved, Passed
    - **Parental Guidance Suggested**: UA / U/A, PG, TV-PG, GP, PG-13
    - **Restricted or Mature Audiences**: A, R, TV-14, TV-MA, 16, Unrated.

```python
# Code
# Creating a function to check logically how the values should be bucketed
def check_certificate_category(certificate):
    if certificate in ['U', 'G', 'Approved', 'Passed']:
        return "For General Audiences"
    elif certificate in ['UA', 'U/A', 'PG', 'TV-PG', 'GP', 'PG-13']:
        return "Parental Guidance Suggested"
    elif certificate in ['A', 'R', 'TV-14', 'TV-MA', '16', 'Unrated']:
        return "For Mature Audiences"
    else:
        return "Not Available"

# Creating a new column Certificate_Category by applying the function to the 'Certificate' column
movies['Certificate_Category'] = movies['Certificate'].apply(check_certificate_category)
```

```python
cc.reset_index(inplace=True)
# When a pivot table is created or group-by operation is performed, the result often has a new index, such as Certificate_Category.
# This means that Certificate_Category is no longer a regular column but an index of the DataFrame.
# When Seaborn tries to access the column Certificate_Category in the  DataFrame cc, it looks for it among the DataFrame's columns.
# Since Certificate_Category is set as the index instead, it cannot be found in the columns, which leads to the ValueError you encountered.
# Hence, reset_index is required.
```

#### Director vs IMDB_Rating

```python
movies.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending = False).head(10)
# movies[movies['Director'],movies['IMDB_Rating'].value_counts() - This won't work as this is a Series method, and not applicable in DataFrames.
```

```python
# Step 1: Group by Star1 and calculate average IMDb rating
avg_rating_star1 = movies.groupby('Star1')['IMDB_Rating'].mean().reset_index()
avg_rating_star1.columns = ['Star1', 'Average_IMDB_Rating']

# Step 2: Count the number of movies by each Star1
star1_counts = movies['Star1'].value_counts().reset_index()
star1_counts.columns = ['Star1', 'Number_of_Movies']

# Step 3: Merge the average ratings and counts
stars_summary = avg_rating_star1.merge(star1_counts, on='Star1', how='left')

# Step 4: Filter stars with 3 or more movies
stars_summary_filtered = stars_summary[stars_summary['Number_of_Movies'] >= 3]

# Step 5: Round the average IMDb rating to 1 decimal point
stars_summary_filtered['Average_IMDB_Rating'] = stars_summary_filtered['Average_IMDB_Rating'].round(1)

# Step 6: Sort by average IMDb rating in descending order
stars_summary_sorted = stars_summary_filtered.sort_values(by='Average_IMDB_Rating', ascending=False).head(10)

# Step 7: Display the result
print(stars_summary_sorted)
```
